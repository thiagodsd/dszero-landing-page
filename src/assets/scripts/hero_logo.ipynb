```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

import seaborn as sns

from IPython.display import display
```

```python
# Mustard Yellow: #FFDB58
# Deep Black: #000000
# Light Gray: #CCCCCC
# Dark Gray: #333333
# Charcoal Gray: #444444
dszero_palette = ["#FFDB58", "#000000", "#CCCCCC", "#333333", "#444444"]
display(sns.color_palette(dszero_palette))

dszero_squash_palette = ["#FFDB58", "#00E0FF", "#FF6DB4"]
display(sns.color_palette(dszero_squash_palette))

dszero_squash_palette_adobe = ["#FFDB58", "#59FFE5", "#FE59FF"]
display(sns.color_palette(dszero_squash_palette_adobe))

dszero_squash_palette_adobe = ["#FFDB58", "#00E0FF", "#FE59FF"]
display(sns.color_palette(dszero_squash_palette_adobe))
```

```python
SEED = 0
```

```python
grid_size = 20
square_size = 0.5
padding = 2*np.sqrt(square_size**2 + square_size**2)
std = 2

fig, ax = plt.subplots(figsize=(10, 10)) 

ax.set_xlim(0, grid_size)
ax.set_ylim(0, grid_size)
ax.set_aspect('equal')

np.random.seed(SEED)
rotations = np.random.randint(0, 90, size=(grid_size, grid_size))

for i in range(grid_size):
    for j in range(grid_size):
        x = i+np.random.normal(0, std)
        y = j+np.random.normal(0, std)
        if x > padding and x < grid_size-padding and y > padding and y < grid_size-padding:
            square = patches.Rectangle((x, y), square_size, square_size, angle=rotations[i, j], fill=True, color=np.random.choice(dszero_squash_palette_adobe))
            ax.add_patch(square)

fig.patch.set_facecolor('#333333')        
ax.axis('off')
plt.savefig("../images/hero_logo_01.png", dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```

```python
grid_size = 20
square_size = 0.5
padding = 2*np.sqrt(square_size**2 + square_size**2)
std = 0

fig, ax = plt.subplots(figsize=(10, 10)) 

ax.set_xlim(0, grid_size)
ax.set_ylim(0, grid_size)
ax.set_aspect('equal')

np.random.seed(0)
rotations = np.random.randint(0, 90, size=(grid_size, grid_size))

COUNT = 0

for i in range(grid_size):
    for j in range(grid_size):
        x = i+np.random.normal(0, std)
        y = j+np.random.normal(0, std)
        if COUNT/(17**2) < 0.33:
            color = dszero_squash_palette_adobe[0]
        elif COUNT/(17**2) < 0.66:
            color = dszero_squash_palette_adobe[1]
        else:
            color = dszero_squash_palette_adobe[2]
        if x > padding and x < grid_size-padding and y > padding and y < grid_size-padding:
            square = patches.Rectangle((x, y), square_size, square_size, fill=True, color=color)
            ax.add_patch(square)
            COUNT += 1

fig.patch.set_facecolor('#333333')        
ax.axis('off')
plt.savefig("../images/hero_logo_03.png", dpi=300, bbox_inches='tight', transparent=True)
plt.show()
```
